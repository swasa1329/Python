```python
# A string is an ordered sequence of characters.

# We can use single quotes or double quotes to represent strings. 

# Multi-line strings can be represented using triple quotes, ”' or “””.
```

#### How to create a string

```python
a='Python'

print("String value :- ", a)
```

```python
a1="Python"

print("String value :- ", a1)
```

```python
a2="""Python"""

print("String value :- ", a2)
```

###### Using str() method :- 

```python
x=str('Welcome')

print("String value :-", x)
```

```python
x=str("""Welcome""")

print("String value :-", x)
```

#### String operations

```python
# We can perform several operations in strings like Concatenation, Repetition, and Slicing
```

#### Concatenation

```python
s1='welcome '
s2='Data Analytics'

print('Conatenation of two strings :-', s1 + s2)
```

#### Repetition

```python
# Repetition: It means repeating a sequence of instructions a certain number of times. 

# "*" operator is a repetition operator for string.
```

```python
s1='welcome '
s2='Data Analytics '

print('Repetition of a string :-', s1 * 3)
print()
print('Repetition of a string :-', s2 * 4)
```
