###### Logical Operator

```python
Logical operators are used for performing AND, OR and NOT operations. 
It either returns True or False according to the condition.

Logical operators are used to combine conditional statements:
To understand the behavior of this logical operator, we can make use of a truth table:
```

```python
# or is True, whenever any (one or more) operand is True.
```

```python
# not will return the opposite of its operand, so True if False is given and vice-versa.
```

```python
x=int(input("Enter the Value of X :- "))

print(x<5 and x<10)

print(x<5 or x<4)

print(not(x<5 and x<4))
```

```python
x = int(input("Enter any value:-"))
print()

print("And Operator-", x > 3 and x < 10)
print("And Operator-", x > 3 and x > 10)
print()

print("Or Operator:-", x > 3 or x < 4)
print("Or Operator:-", x > 30 or x < 14)
print()

print("Not Operator:-", not(x > 3 and x < 10))
print("Not Operator:-", not(x > 10 and x < 15))
print()
print("Not Operator:-", not(x > 3 or x < 10))
print("Not Operator:-", not(x > 10 or x < 15))
print()
```

###### Identity Operators

```python
Identity operators are used to compare the objects, not if they are equal,
but if they are actually the same object, with the same memory location.

Identity operators check whether two values are identical. Python has 2 identity operators as well:

is
is not
```

###### is Operator

```python
# The is operator returns True if the first value is the same as the second. Otherwise, it returns False.
# It returns True, if two objects have the same identity.

x=3
y=3.0
print("is Operator:- ", x is y)

x=30
y=30

print("is Operator:- ", x is y)

x="Abode"
y="AAbode"

print("is Operator:- ", x is y)

x=50
x=y
print("is Operator:- ", x is y)
```

```python
# The is not operator returns True if the first value is not identical to the second. It returns False otherwise.

x=3
y=3.0
print("is not Operator:- ", x is not y)

x=30
y=30

print("is not Operator:- ", x is not y)

x="Abode"
y="AAbode"

print("is not Operator:- ", x is not y)

x=50
# x=y
y=x
print("is not Operator:- ", x is not y)
```

```python
x1=5
y1=5

print("is Operator:- ", x1 is y1)
print("is not Operator:- ", x1 is not y1)

x2="Hello"
y2="Hello"
print()

print("is Operator:- ", x2 is y2)
print("is not Operator:- ", x2 is not y2)
print()

x3=[1,2,3,4] #list operator memory location is different when compared to veriable memory
y3=[1,2,3,4]

print("is Operator:- ", x3 is y3)
print("is not Operator:- ", x3 is not y3)

# x1 and y1 are integers of the same values, so they are equal as well as identical.
# # Same is the case with x2 and y2 (strings).

# # But x3 and y3 are lists. They are equal but not identical. 
# # It is because the interpreter locates them separately in memory although they are equal.
```

```python
x3=[1,2,3,4]
y3=x3

print("is Operator:- ", x3 is y3)
```

###### Membership Operators

```python
# These operators test whether a value is a member of a sequence. The sequence may be a list, a string, or a tuple.
# We have two membership python operators- in and not in.
```

```python
# in Operator in Python:-
# This checks if a value is a member of a sequence.

pets=['dog','cat','ferret']

print("in Operator :- ", 'cat' in pets)
print("in Operator :- ", 'fox' in pets)

x=[1,2,3,4,5]

print("in Operator :- ", 3 in x)
print("in Operator :- ", 6 in x)

print("in operator:- ", 'me' in 'disappointment')
print()
print("in operator:- ", 'Help' in 'Holddisappointment')
```

```python
# not in Operator in Python:-
# Unlike in, not in checks if a value is not a member of a sequence.

pets=['dog','cat','ferret']
print("not in Operator :- ", 'cat' not in pets)
print("not in Operator :- ", 'fox' not in pets)

x=[1,2,3,4,5]
print("not in Operator :- ", 3 not in x)
print("not in Operator :- ", 6 not in x)

print("not in operator:- ", 'me' not in 'disappointment')
print()
print("not in operator:- ", 'Help' not in 'Holddisappointment')
```

###### BITWISE LEFTSHIFT Operator

```python
Python bitwise left shift operator shifts the left operand bits towards the left side 
for the given number of times in the right operand.
In simple terms, the binary number is appended with 0s at the end.
```

```python
A=9

print("The binary value of A is:-", bin(A))

print("Bitwise Operation - Left Shift with 2 :-", A<<2)

x = 0b100100
print("The integer value of x is:-",int(x))
```

```python
A=9

print("The binary value of A is:-", bin(A))

print("Left Shift with 2 :-", A<<3)

x = 0b1001000
print("The integer value of x is:-",int(x))
```

###### BITWISE RIGHTSHIFT Operator

```python
Python right shift operator is exactly the opposite of the left shift operator. 
Then left side operand bits are moved towards the right side for the given number of times.
In simple terms, the right side bits are removed.
```

```python
A=9
B=2
print(bin(A))
print(A>>B)
print(int(0b0010))
```
