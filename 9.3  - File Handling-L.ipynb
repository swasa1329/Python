###### What are Python files?

```python
If you are working in a large software application where they process a large number of data, 
then we cannot expect those data to be stored in a variable as the variables are volatile in nature.

Hence when are you about to handle such situations, the role of files will come into the picture.

As files are non-volatile in nature, the data will be stored permanently in a secondary device
like Hard Disk and using python we will handle these files in our applications.
```

```python
Files are data or information that exist in the computer’s storage devices. 
These units of information are used to permanently keep data on a hard disk. 
By using files, we can keep data for future use.

A file is a named location on the disk which is used to store the data permanently.

As a computer user, you certainly know about music files, video files, and text files. 
Python allows you to manipulate these files.
```

###### Types of file in Python

```python
There are two types of files in Python and each of them are explained below in detail with examples for your easy understanding.

They are:

1. Binary file

2. Text file
```

###### 1. Binary files in Python

**Most of the files that we see in our computer system are called binary files.**

Example:

**1. Document files:** .pdf, .doc, .xls etc.

**2. Image files:** .png, .jpg, .gif, .bmp etc.

**3. Video files:** .mp4, .3gp, .mkv, .avi etc.

**4. Audio files:** .mp3, .wav, .mka, .aac etc.

**5. Database files:** .mdb, .accde, .frm, .sqlite etc.

**6. Archive files:** .zip, .rar, .iso, .7z etc.

**7. Executable files:** .exe, .dll, .class etc.

```python
All binary files follow a specific format. 

We can open some binary files in the normal text editor, but we can’t read the content present inside the file. 
That’s because all the binary files will be encoded in the binary format,
which can be understood only by a computer or machine.

For handling such binary files, we need a specific type of software to open it.

For Example, You need Microsoft word software to open .doc binary files. 

Likewise, you need a pdf reader software to open .pdf binary files and 
you need a photo editor software to read the image files and so on.
```

###### Text files in Python

Text files don’t have any specific encoding and it can be opened in normal text editor itself.

Example:

**1.Web standards:** html, XML, CSS, JSON etc.

**2.Source code:** c, app, js, py, java etc.

**3.Documents:** txt, tex, RTF etc.

**4.Tabular data:** csv, tsv etc.

**5.Configuration:** ini, cfg, reg etc.

###### What is file handling in Python?

```python
Having files is not enough — we need to be able to manipulate them either by opening, closing, writing, or deleting them.

File handling in Python is the process of using Python to carry out the following operations on files:

1. Create a file 
2. Open a file
3. Read file
4. Write to file
5. Append to file
6. Close a file
7. Rename files
8. Delete files
```

###### 2. Open a File

Before starting any of that, we must first open the files in the right mode. 

There are three modes available, namely, read, write, or append.

1. Read, which is written as “r” in Python, allows you to read the file but doesn’t let you edit or delete any content.

2. Write, denoted by “w”, means that if the file exists, all content is deleted before it’s opened to write.

3. Append, “a”, opens the file in append mode.

#### Open A File

```python
While using binary files, we have to use the same modes with the letter ‘b’ at the end.

So that Python can understand that we are interacting with binary files.
```

```python
# To open a file for reading, it is enough to specify the name of the file.

# Assume we have the following file, located in the same folder:-

f1= open("Sales_2.txt", 'r') # To Opening a file we will use this

print(f1.read())
```

###### 3. Read the data from File

```python
In order to read a file in python, we must open the file in read mode.

There are three ways in which we can read the files in python.

1. read([n])
2. readline([n])
3. readlines()

Here, n is the number of bytes to be read.
```

```python
# In order to read the file, first, we need to open the file in reading mode.

# Will not support files,that has no extensions.
```

```python
# Example 1:- 

f1= open("sales.txt", "r") # To Opening a file, we will use the following way

print("The Contents of the file:- \n",  f1.read(5)) 

f1.close

# Output: Hello

# Here we are opening the file test.txt in a read-only mode and 
# are reading only the first 5 characters of the file using the f1.read(5) method.
```

##### Reading a File Line by Line

```python
# By looping through the lines of the file, you can read the whole file, line by line:

# Example:- Loop through the file line by line:

f = open("sales.txt", "r")

for x in f:
    print(x)
```

##### 4. Writing Data to a File

```python
In order to write data into a file, we must open the file in write mode.

We need to be very careful while writing data into the file as
it overwrites the content present inside the file that you are writing,
and all the previous data will be erased.

We will use the below method for writing data into a file as shown below.

1. write(string)
```

**Note:** the "w" method will overwrite the entire file.

```python
# Example:-1

file1 = open("sales.txt", 'w')

file1.write("Hello World")

# The above code writes the String ‘Hello World’ into the ‘sales.txt’ file.
```

```python
# Example:-2

file2 = open("sales.txt", "w")

file2.write("Hello World-1 \n")

file2.write("Hello Python-1")

# The first line will be ‘Hello World-1’ and as we have mentioned \n character,
# the cursor will move to the next line of the file and then write ‘Hello Python-1’.

# Remember if we don’t mention \n character, then the data will be written continuously 
# in the text file like ‘Hello WorldHelloPython’
```

```python
# Example:-3

f2= open("sales.txt", "w")

f2.write("Adding a new content to the file")

f2.close()

# open and read the file after the overwriting:

f2= open("sales.txt", "r")

print(f2.read())
```

```python
# Write to an Existing File

- To write to an existing file, you must add a parameter to the open() function:

    "a" - Append - will append to the end of the file

    "w" - Write - will overwrite any existing content
```

###### 5. Append()

```python
# To append data into a file we must open the file in ‘a’ or 'a+' mode.

# Example 1:
    
my_file = open("sales.txt", "a")

my_file.write ("Strawberry")

# The above code appends the string ‘Strawberry’ at the end of the ‘sales.txt’ file.
```

```python
# Example 2:

f = open("sales_1.txt", "a")

f.write("\n Now the file has more content!")

f.close()

# open and read the file after the appending:

f = open("sales_1.txt", "r")

print(f.read())
```

###### 6. Python Rename() method

```python
Python provides us with an “os” module which has some in-built methods
that would help us in performing the file operations such as renaming and deleting the file.

In order to use this module, first of all, we need to import the “os” module 
in our program and then call the related methods.
```

```python
This rename() method accepts two arguments i.e. the current file name and the new file name.

Syntax:- os.rename(current_file_name, new_file_name)
```

```python
# Example 1:

import os

os.rename("sales_1.txt", "sales1.txt")

# Here ‘sales_1.txt’ is the current file name and ‘sales1.txt’ is the new file name.
```

###### 7. remove() method:

```python
We use the remove() method to delete the file by supplying 
the file name or the file location that you want to delete.

Syntax:- os.remove(file_name)
```

```python
# Example 1:

import os  # we need to import a lib as "os" for delete a file and use below method as os.remove()

os.remove("sales.txt")

# Here "sales.txt" is the file that you want to remove.
```
